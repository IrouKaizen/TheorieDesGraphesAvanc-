```python
import networkx as nx
import matplotlib.pyplot as plt


def draw_graph(G, pos, edge_labels, path_edges=None):
    plt.figure(figsize=(10, 6))
    nx.draw_networkx_nodes(G, pos, node_size=500, node_color='blue')
    nx.draw_networkx_edges(G, pos, width=1, style='dashed')
    nx.draw_networkx_labels(G, pos, font_size=12, font_family="sans-serif")
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    if path_edges:
        nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=2)
    plt.axis('on')
    plt.show()


G = nx.Graph()
edges = [
    ('San Francisco', 'Los Angeles', 39),
    ('San Francisco', 'Denver', 55),
    ('San Francisco', 'Chicago', 99),
    ('San Francisco', 'New York', 129),
    ('Los Angeles', 'Denver', 89),
    ('Los Angeles', 'New York', 129),
    ('Denver', 'Chicago', 69),
    ('Chicago', 'Boston', 79),
    ('Chicago', 'New York', 65),
    ('Chicago', 'Atlanta', 99),
    ('Boston', 'New York', 39),
    ('New York', 'Atlanta', 79),
    ('New York', 'New Jersey', 89),
    ('Atlanta', 'New Jersey', 89),
]

G.add_weighted_edges_from(edges)
pos = nx.spring_layout(G)
edge_labels = nx.get_edge_attributes(G, 'weight')
draw_graph(G, pos, edge_labels)

while True:
    start_node = input("Entrez la ville de dpart ou 'quitter' pour sortir: ")
    if start_node.lower() == 'quitter':
        break
    end_node = input("Entrez la ville d'arrive ou 'quitter' pour sortir: ")
    if end_node.lower() == 'quitter':
        break

    if start_node in G and end_node in G:
        shortest_path = nx.dijkstra_path(G, source=start_node, target=end_node)
        path_edges = list(zip(shortest_path, shortest_path[1:]))
        draw_graph(G, pos, edge_labels, path_edges)
    else:
        print("L'une des villes entrs n'existe pas dans le graphe. Veuillez ressayer avec des noms de villes valides.")
        continue_input = input("Voulez-vous ressayer? Entrez 'oui' pour continuer ou 'quitter' pour sortir: ")
        if continue_input.lower() == 'quitter':
            break
```
